```python
import matplotlib.pyplot as plt
import pandas as pd
category_total=df.groupby('CATEGORY')['AMOUNT'].sum()
category_total.plot(kind='bar')
plt.xlabel('CATEGORY')
plt.ylabel('TOTAL AMOUNT')
plt.title('EXPENSES BY CATEGORY')
plt.show()
```

```python
import pandas as pd
df=pd.read_csv('expenses.csv')
print(df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
totalamt=df.groupby('CATEGORY' , as_index=False)['AMOUNT'].sum()
sns.barplot(x='CATEGORY',y='AMOUNT',data=totalamt)
plt.xlabel('CATEGORY')
plt.ylabel('TOTAL AMOUNT')
plt.title('EXPENSES BY CATEGORY')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.histplot(df['AMOUNT'])

plt.xlabel('CATEGORY')
plt.ylabel('TOTAL AMOUNT')
plt.title('EXPENSES BY CATEGORY')
plt.show()
```
